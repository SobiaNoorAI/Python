**drop_duplicates():** drop duplicate values

```python
import pandas as pd

xlsx_url = r'https://raw.githubusercontent.com/SobiaNoorAI/Python/main/Pandas%20/student_result.xlsx'

#open excel file
student_result = pd.read_excel(xlsx_url, sheet_name='Sheet1')

#drop duplicate data
student_result.drop_duplicates(inplace = True)
print("\n Drop the duplicate data:")
student_result
```

Now, duplicate data of Nigel has been removed.

# Grouping and Sorting Data

**Groupby():** Divide our data into groups according to specific columns

```python
# group data based on Grade column
grade_avg_scores = student_result.groupby('Grade')['Score'].median()

print("\nAverage Score by Grade:")
grade_avg_scores
```

Here, we make a group of the data according to grade and then find the median value.
